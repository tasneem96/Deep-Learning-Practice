```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/deep-learning-keras-tf-tutorial/master/13_dropout_layer/sonar_dataset.csv',header = None)
df
```

```python
 df.shape
```

```python
df.isna().sum()
```

```python
df[60].value_counts()
```

```python
df1 = pd.get_dummies(data = df, columns=[60])
```

```python
df1.drop('60_M',axis = 'columns',inplace =True)
```

```python
df1.shape
```

```python
df1
```

```python
df1['60_R'].value_counts()
```

```python
x = df1.drop('60_R',axis = 1)
y = df1['60_R']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test ,y_train,y_test = train_test_split(x,y,test_size = .2, random_state = 5)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from threading import active_count
model = keras.Sequential(
    [
     keras.layers.Dense(60,input_shape = (60,),activation = 'relu'),
     keras.layers.Dropout(.5),
     keras.layers.Dense(30,activation = 'relu'),
     keras.layers.Dropout(.5),
     keras.layers.Dense(15,activation = 'relu'),
     keras.layers.Dropout(.5),
     keras.layers.Dense(1,activation = 'sigmoid'),
    ]
)

model.compile(
    optimizer = 'adam',
    loss = keras.losses.BinaryCrossentropy(from_logits = True),
    metrics = ['accuracy']
)
```

```python
model.fit(x_train,y_train,epochs = 100, batch_size = 5)
```

```python
y_pred = model.predict(x_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
model.evaluate(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```
