```python
def sigmoid_np(x):
  import math
  return 1/(1+np.exp(-x))
```

```python
def log_loss1 (y_true,y_predicted):
  epsilon = 1e-15
  new_y_predicted = [max(i,epsilon) for i in y_predicted]
  new_y_predicted = [min (i,1-epsilon) for i in new_y_predicted]
  new_y_predicted = np.array(new_y_predicted)
  return -np.mean(y_true * np.log(new_y_predicted) + (1-y_true)* np.log(1-new_y_predicted))
```

```python
def gradient_descent(Age,Affordibility,y_true,epoch):
  w1 = w2 = 1
  b = 0
  rate = 0.5
  n = len(Age)

  for i in range(epoch):
    weighted_sum = w1*Age + w2*Affordibility + b
    y_predicted = sigmoid_np(weighted_sum)

    loss = log_loss1(y_true , y_predicted)

    w1d = (np.dot(np.transpose(Age),(y_predicted - y_true)))/n
    w2d = (np.dot(np.transpose(Affordibility),(y_predicted - y_true)))/n
    bd = np.mean(y_predicted - y_true)

    w1 = w1-rate*w1d
    w2 = w2-rate*w2d
    b = b-rate*bd

    print (f'Epoch {i} , w1 {w1}, w2 {w2} , loss {loss}')

  return (w1, w2, b)
```

```python
gradient_descent(x_train_scaled['Age'], x_train_scaled ['Affordibility'], y_train, 1000)
```
